```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

from google.colab import files
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]
image = cv2.imread(filename) #opencv read in BGR format
cv2_imshow(image) # show the original image
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
I = [image,gray]
for i in I:
    resized = cv2.resize(i,(200,200))
    cv2_imshow(resized)
```

```python
annoted = image.copy()
cv2.rectangle(annoted,(300,200),(650,650),(0,0,255),10)
cv2.putText(annoted,'My.box',(10,55),cv2.FONT_HERSHEY_SIMPLEX,1,(0,255,0),2)
cv2_imshow(annoted)
```

```python
cv2.imwrite('output_image.jpg',annoted)
print("Image saved as output_image.jpg")
```

```python
#Example of data augmentation

from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img, img_to_array
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
img = load_img(filename, target_size=(224,224))
img_array = img_to_array(img)
img_array = np.expand_dims(img_array, axis =0)


datagen = ImageDataGenerator(
    rotation_range=70,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range = [0.5,1.5]
)
i = 0
for batch in datagen.flow(img_array,batch_size=1):
    plt.figure(figsize = (2,2))
    plt.imshow(batch[0].astype('uint8'))
    plt.axis('off')
    i += 1
    if i > 5:
          break
```
